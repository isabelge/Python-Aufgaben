## Aufgabe 1
Das Schlüsselwort `if` wird in Python verwendet um eine Verzweigung zu schreiben. Nach einem `if` kommt eine Bedingung und abhängig von dieser Bedingung wird der Code in der Schleife ausgeführt oder nicht.

### Übersicht Operatoren
| Zeichen | Bedeutung      | Beispiel |   |   |
|---------|----------------|----------|---|---|
| ==      | gleich         | x == y   |   |   |
| !=      | ungleich       | x != y   |   |   |
| >       | grösser        | x > y    |   |   |
| >=      | grösser gleich | x >= y   |   |   |
| <       | kleiner        | x < y    |   |   |
| <=      | kleiner gleich | x <= y   |   |   |

Die erste Aufgabe besteht darin, die korrekte Bedingung in die Verzweigung einzusetzen. Nehmen Sie dafür das Struktogramm aus der Einstiegsaufgabe zur Hilfe.

````python
## die Punkte können angepasst werden, um den Code zu testen
punkte = 12

## nach dem if und vor dem `:` muss der korrekte Operator eingefügt werden
if punkte > 10:
    print("Prüfung bestanden")
else:
    print("Prüfung nicht bestanden")
````

## Aufgabe 5
Auch in dieser Aufgabe geht es darum, die vollständige Verzweigung zu schreiben. Es gibt nun allerdings zusätzliche Logik.

In diesem Programm muss der Durchschnitt über mehrere Noten berechnet werden. Zum Schluss soll das Programm zurückgeben, ob der Notendurchschnitt ungenügend oder genügend ist.

````python
note1 = 4
note2 = 6
note3 = 2

## hier muss der Durchschnitt berechnet werden
durchschnitt = (note1 + note2 + note3) / 3

## hier muss die Verzweigung mit if und else hinzugefügt werden
if durchschnitt >= 4:
    print("Der Notendurchschnitt ist genügend.")
else:
    print("Der Notendurchschnitt ist ungenügend.")
````

## Aufgabe 6
Der Code aus der 5. Aufgabe soll nun noch erweitert werden. Am Schluss soll nicht nur ausgegeben werden ob der Notendurchschnitt `ungenügend` oder `genügend` war, sondern zusätzlich auch noch bei `genügend` zwischen `gut` und `sehr gut` unterscheiden.

| Bedingung              | Notenprädikat |
|------------------------|---------------|
| Durchschnitt < 4       | ungenügend    |
| 4 ≤ Durchschnitt < 4.5 | genügend      |
| 4.5 ≤ Durchschnitt < 5 | gut           |
| 5 ≤ Durchschnitt ≤ 6   | sehr gut      |

Damit ein Bereich zwischen zwei Zahlen in einem `if` angegeben werden kann, wird eine verkettete Vergleichsoperation verwendet.

Zum Beispiel: `4 <= durchschnitt < 4.5`

Ist der Wert vom Durchschnitt grösser oder gleich 4 und gleichzeitig kleiner als 4.5?


````python
note1 = 3
note2 = 5
note3 = 4

# Durchschnitt berechnen
durchschnitt = (note1 + note2 + note3) / 3

if durchschnitt < 4:
    print("ungenügend")
elif 4 <= durchschnitt < 4.5:
    print("genügend")
elif 4.5 <= durchschnitt < 5:
    print("gut")
elif 5 <= durchschnitt <= 6:
    print("sehr gut")
else:
    print("Ungültiger Wert (Durchschnitt ausserhalb 1–6)")
````

Wie haben sie die Aufgabe 6 gelöst? Haben sie mehrere `if` Anweisungen verwendet?

Haben sie die Aufgabe so gelöst?

```
note1 = 3
note2 = 5
note3 = 4

# Durchschnitt berechnen
durchschnitt = (note1 + note2 + note3) / 3

if durchschnitt < 4:
    print("ungenügend")
if 4 <= durchschnitt < 4.5:
    print("genügend")
if 4.5 <= durchschnitt < 5:
    print("gut")
if 5 <= durchschnitt <= 6:
    print("sehr gut")
else:
    print("Ungültiger Wert (Durchschnitt ausserhalb 1–6)")
```

Anstatt mehrere `if` zu verwenden, gibt es auch die Möglichkeit, `elif` zu nutzen. `elif` gehört zum gleichen Block wie das vorherige `if` und kann nicht alleine stehen. Es wird nur dann geprüft, wenn alle vorherigen Bedingungen falsch waren.

Das bedeutet: Sobald ein `elif` wahr ist, werden alle weiteren `elif` und `else` übersprungen.

Wenn mehrere `if` nacheinander verwendet werden, wird jede einzelne Bedingung geprüft.

- `if` → startet einen neuen Verzweigungsblock
- `elif` → weitere Abfrage, aber nur dann, wenn die vorherige Bedingung nicht erfüllt wurde
- `else` → wird ausgeführt, wenn alle vorherigen Bedingungen nicht erfüllt wurden

## Aufgabe 7

Schreiben Sie ein Programm, das die Alterskategorie bestimmen kann. Verwenden Sie dazu `if`, `elif` und `else`.

- Alter < 12 → Kind
- Alter zwischen 12 und 18 → Teenager
- Alter zwischen 18 und 64 → Erwachsen
- Alter > 65 → Senior

````python
alter = 30

## hier muss die Verzweigung mit if, elif und else hinzugefügt werden

if alter < 12:
    print("Alterskategorie: Kind")
elif 12 >= alter < 18:
    print("Alterskategorie: Teenager")
elif 18 >= alter < 64:
    print("Alterskategorie: Erwachsen")
else:
    print("Alterskategorie: Senior")
````

## Aufgabe 8

Die Aufgabe besteht darin, die Software eines Getränkeautomaten zu schreiben. Der Getränkeautomat hat folgende Funktionalitäten:

- Es gibt mehrere Getränke zur Auswahl: Wasser, Limonade und Cola.
- Wasser kostet 1 CHF, Limonade 2 CHF und Cola 5 CHF.
- Die Getränke werden jeweils mit nur einem Geldstück bezahlt.
- Das Festlegen des zu zahlenden Betrags und das Überprüfen der Zahlung muss mit einer `if`-`elif`-`else`-Anweisung erfolgen.


````python
auswahl = int(input("Bitte wählen Sie Ihr Getränk aus: 1 für Wasser, 2 für Limonade oder 3 für Cola"))

## Je nachdem, welches Getränk gewählt wurde, muss der korrekte Preis angegeben werden
kosten = 0
if auswahl == 1:
    kosten = 1
    print("Wasser: 1.00 CHF")
elif auswahl == 2:
    kosten = 2
    print("Limonade: 2.00 CHF")
elif auswahl == 3:
    kosten = 5
    print("Cola: 5.00 CHF")

geld = int(input("Bitte geben Sie den Betrag passend ein"))

## Nun muss der Betrag überprüft werden und entweder erscheint die Meldung
## "Vielen Dank, bitte entnehmen Sie Ihr Getränk"
## oder
## "Der Betrag wurde nicht korrekt eingeworfen."

if geld == kosten:
    print("Vielen Dank, bitte entnehmen Sie Ihr Getränk")
else:
    print("Der Betrag wurde nicht korrekt eingeworfen.")
````

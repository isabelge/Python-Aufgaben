### Teil 4
Die Bank hat entschieden, eine Gebühr einzuführen. Immer wenn Geld gewechselt wird, muss der Bank 1 Franken bzw. 1 Euro bezahlt werden.
Passen Sie den Code entsprechend an. Es muss jeweils ein zweiter Wert an die Funktion übergeben werden und diese Gebühren müssen vor der Umrechnung abgezogen werden. Ändern Sie den Code so, dass die Gebühren korrekt berücksichtigt werden.

````python

def franken_euro (a):
    return a * 1.09

def euro_franken (a):
    return a * 0.92

eingabe = int(input("Bitte wählen Sie für die Umrechnung von Franken zu Euro die 1 und für die Umrechnung von Euro zu Franken die 2"))

gebuehren = 1

if eingabe == 1:
    franken = int(input("Bitte geben Sie den Betrag in Franken ein: "))
    print(franken_euro(franken))
elif eingabe == 2:
    euro = int(input("Bitte geben Sie den Betrag in Euro ein: "))
    print(euro_franken(euro))
else:
    print("Ungültige Eingabe!")
````

````python
#### Lösung ####
def franken_euro (a, gebuehren):
    return (a - gebuehren) * 1.09

def euro_franken (a, gebuehren):
    return (a - gebuehren) * 0.92

eingabe = int(input("Bitte wählen Sie für die Umrechnung von Franken zu Euro die 1 und für die Umrechnung von Euro zu Franken die 2"))

gebuehren = 1

if eingabe == 1:
    franken = int(input("Bitte geben Sie den Betrag in Franken ein: "))
    print(franken_euro(franken, gebuehren))
elif eingabe == 2:
    euro = int(input("Bitte geben Sie den Betrag in Euro ein: "))
    print(euro_franken(euro, gebuehren))
else:
    print("Ungültige Eingabe!")
````

````python
### Lösung ###

def berechnung_preis(original_preis, rabatt):
    return original_preis * (1 - rabatt / 100)

original_preis = int(input("Bitte geben Sie den original Preis in Franken ein: "))

rabatt_preis = int(input("Bitte geben Sie den Rabatt in Franken ein: "))

neuer_preis = berechnung_preis(original_preis, rabatt_preis)
print(neuer_preis)
````
